```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load cleaned dataset
df_cleaned = pd.read_csv('/Users/admin/My Projects/ecommerce-advertising/data/ecommerce_cleaned_data.csv')
```

```python
# Example visualizations

# 1. Bar chart: Product Categories by Total Purchases
category_sales = df_cleaned.groupby('Product Category')['Total Purchase Amount'].sum()
category_sales.plot(kind='bar', title='Total Purchases by Category')
plt.show()
```

```python
# 2. Pie chart: Payment Method Distribution
payment_method = df_cleaned['Payment Method'].value_counts()
payment_method.plot(kind='pie', title='Payment Method Distribution', autopct='%1.1f%%')
plt.show()
```

```python
# 3. Histogram: Customer Ages
df_cleaned['Customer Age'].plot(kind='hist', bins=10, title='Customer Age Distribution')
plt.show()
```

```python
# 4. Line chart: Purchases Over Time
# Convert 'Purchase Date' to datetime
df_cleaned['Purchase Date'] = pd.to_datetime(df_cleaned['Purchase Date'])

# Group by month and sum the 'Total Purchase Amount'
purchases_over_time = df_cleaned.groupby(df_cleaned['Purchase Date'].dt.to_period('M'))['Total Purchase Amount'].sum()

# Plot the data
purchases_over_time.plot(kind='line', title='Purchases Over Time')
plt.ylabel('Total Purchase Amount')
plt.xlabel('Month')
plt.show()
```
